<div class = "alert alert-block alert-success" style="border-radius: 20px;"><h2 style="font-size:40px">2.</h2>This is our get_SmartArray_LogicalDrives function overview.<br>
    <i>Let me give you a walk through of it!</i>
    <br>
There is a hierarchy involved in iLO Rest Api. So we are simply following the hierarchy present for error free execution.
    <br>
    
    
1. The variables smartstorage_response and smartarraycontrollers are initialized as an empty list and dictionary, respectively.
    <br>

2. The get_resource_directory function is used in the code to try and retrieve the resource directory. The code resorts to a manual search for the pertinent URIs if the resource directory is unavailable or the <b>DISABLE_RESOURCE_DIR </b>flag is set to <b>True</b>.The code repeats the preceding code snippet's procedures to retrieve the pertinent URIs for smart storage and array controllers if the resource directory is not in use or unavailable.
    <br>

3. The code searches through each instance in the resource directory for instances of the specified type if the resource directory is used. It fetches the response, which is an array of array controller members, and extracts the instance's URI. After locating the first pertinent instance, the loop ends.
The smartstorage_response is then used by the code, presuming it contains the array of array controller elements.
The code obtains specific information about each controller in smartstorage_response and stores it in the smartarraycontrollers dictionary with the controller's @odata.id as the key.
    <br>
    
4.For the specified smart storage array controller, it prints the logical drive URIs.The code gets the logical drives URI from the controller's links, sends an HTTP GET request to get the answer, and sets the result to the logicaldrives_resp variable. 
    <br>
    
5.The <b>LogicalDrives</b> key in the smartarraycontrollers dictionary is used to add the logical drive members.
A message is sent to stderr if there are no logical drives.
  <br>
    
6.The code publishes information about each logical drive in <b>logicaldrives_resp</b>, fetches detailed drive data with an HTTP GET request, and prints the drive data as JSON.
 <br>
    
7. The initial values of the variables smartarraycontrollers and smartstorage_response are a dictionary and an empty 
list respectively.The code configures the smartstorage_uri to point to the endpoint 
<b>/redfish/v1/Systems/1/SmartStorage/ArrayControllers/</b> for Gen9 servers. To obtain data about the array
controllers, use this endpoint.
    <br>

8. An HTTP GET request is sent to the smartstorage_uri, and the array of array controller members is extracted 
from the response to populate the smartstorage_response.The code obtains specific information about each controller 
in smartstorage_response and stores it in the smartarraycontrollers dictionary with the controller
@odata.id as the key.
    <br>
9.The logical drive URIs for the specific smart storage array controller are printed by the code.The controller links, 
specifically the LogicalDrives key under the Links object, are where the logicaldrives_uri is located.In order to 
receive the logical drives response, which provides details about the logical drives connected to the controller, 
the code performs an HTTP GET request to the logicaldrives_uri.
<br>
10.If the number of logical drive members is zero, an error message stating that no logical 
drives are accessible for this controller is sent to stderr.Otherwise, the code publishes information about each 
logical drive in the logicaldrives_resp, makes an HTTP GET request to the logical drive URI ,obtains detailed drive 
data, and prints the drive data as JSON.

 </div>

```python
def get_SmartArray_LogicalDrives_gen9(_redfishobj):

    smartstorage_response = []
    smartarraycontrollers = dict()

    smartstorage_uri = "/redfish/v1/Systems/1/SmartStorage/ArrayControllers/"
    print (smartstorage_uri)
    smartstorage_response = _redfishobj.get(smartstorage_uri).obj['Members']

    for controller in smartstorage_response:
        smartarraycontrollers[controller['@odata.id']] = _redfishobj.get(controller['@odata.id']).obj
        sys.stdout.write("Logical Drive URIs for Smart Storage Array Controller \'%s\' : \n" \
                                        % smartarraycontrollers[controller['@odata.id']].get('Id'))
        logicaldrives_uri = smartarraycontrollers[controller['@odata.id']].Links\
                                                                    ['LogicalDrives']['@odata.id']
        print (logicaldrives_uri)
        logicaldrives_resp = _redfishobj.get(logicaldrives_uri)
        if logicaldrives_resp.dict['Members@odata.count'] == 0:
            sys.stderr.write("\tLogical drives are not available for this controller.\n")
        else:
            for lds in logicaldrives_resp.dict['Members']:
                sys.stdout.write("\t An associated logical drive: %s\n" % logicaldrives_resp.dict['Name'])
                drive_data = _redfishobj.get(lds['@odata.id']).dict
                print(json.dumps(drive_data, indent=4, sort_keys=True))
def get_SmartArray_LogicalDrives_gen9(_redfishobj):

    smartstorage_response = []
    smartarraycontrollers = dict()

    smartstorage_uri = "/redfish/v1/Systems/1/SmartStorage/ArrayControllers/"
    print (smartstorage_uri)
    smartstorage_response = _redfishobj.get(smartstorage_uri).obj['Members']

    for controller in smartstorage_response:
        smartarraycontrollers[controller['@odata.id']] = _redfishobj.get(controller['@odata.id']).obj
        sys.stdout.write("Logical Drive URIs for Smart Storage Array Controller \'%s\' : \n" \
                                        % smartarraycontrollers[controller['@odata.id']].get('Id'))
        logicaldrives_uri = smartarraycontrollers[controller['@odata.id']].Links\
                                                                    ['LogicalDrives']['@odata.id']
        print (logicaldrives_uri)
        logicaldrives_resp = _redfishobj.get(logicaldrives_uri)
        if logicaldrives_resp.dict['Members@odata.count'] == 0:
            sys.stderr.write("\tLogical drives are not available for this controller.\n")
        else:
            for lds in logicaldrives_resp.dict['Members']:
                sys.stdout.write("\t An associated logical drive: %s\n" % logicaldrives_resp.dict['Name'])
                drive_data = _redfishobj.get(lds['@odata.id']).dict
                print(json.dumps(drive_data, indent=4, sort_keys=True))
```

<div class="alert alert-block alert-success" style="border-radius: 20px;"><h2 style="font-size:40px">3.</h2>
<b>The Main function: </b> <br>
    <ul>
        <li>An instance of the argparse is created by the code.assigns the ArgumentParser class to the parser variable. The goal of the script is described in the description parameter.</li>
        <li> Using the add_argument method of the ArgumentParser class, the code adds the following three command-line arguments:1.<b>-i</b> or --ilo: Specifies the iLO IP address of the server<br>2.<b>u</b> or --user: Indicates the iLO login username. It has the dest, action, help, and default parameters, just as the -i argument.<br>
            3.<b>-p</b> or --password: Specifies the iLO password for login.
        </li>
        <li>A new variable named system_url is created and given the value "https://" + options.ilo_ip by the code. This creates the system's URL using the supplied iLO IP address. The iLO IP address will come after the "https://" in the final URL.The created URL will be shown on the console once the code outputs the system_url variable.</li>
        <li> The DISABLE_RESOURCE_DIR option is set to False by the code. This option is used to specify whether or not the resource directory should be disabled, along with any related procedures. It is specifically set to False in this instance, allowing the resource directory to be utilised.</li>
        <li>The base_url, username, and password that were received through the command-line inputs are passed to the code, which then creates a RedfishClient object with the name redfish_obj.The specified credentials are used to log in to the server by calling the redfish_obj.login() method. A session is established with the server as a result.A ServerDownOrUnreachableError exception is thrown if the server cannot be reached or does not support Redfish, and the script is terminated after printing an error message to stderr.</li>
        <li>The redfish_obj is supplied as an argument when the get_gen() function is used. The server generation data is retrieved using this function.The second value of the tuple given by get_gen() is disregarded using , and the server generation is kept in the ilogen variable.The console is printed with the server generation. </li>
        <li>The get_SmartArray_LogicalDrives() function is called with the redfish_obj parameter to obtain details on the smart array logical drives if the server generation is equal to 5, indicating a Gen5 server.
The get_SmartArray_LogicalDrives_gen9() function is called with the redfish_obj parameter to obtain details about the smart array logical drives unique to Gen9 servers if the server generation is not 5, indicating a different generation.</li>
        <li>Finally logout from the session can be done by using redfish_obj.logout()</li>
    </ul>
</div>

```python
if __name__ == "__main__":

    # Initialize parser
    parser = argparse.ArgumentParser(description = "Script to upload and flash NVMe FW")

    parser.add_argument(
        '-i',
        '--ilo',
        dest='ilo_ip',
        action="store",
        help="iLO IP of the server",
        default=None)
    parser.add_argument(
        '-u',
        '--user',
        dest='ilo_user',
        action="store",
        help="iLO username to login",
        default=None)
    parser.add_argument(
        '-p',
        '--password',
        dest='ilo_pass',
        action="store",
        help="iLO password to log in.",
        default=None)

    options = parser.parse_args()

    system_url = "https://" + options.ilo_ip
    print (system_url)

    # flag to force disable resource directory. Resource directory and associated operations are
    # intended for HPE servers.
    DISABLE_RESOURCE_DIR = False

    try:
        # Create a Redfish client object
        redfish_obj = RedfishClient(base_url=system_url, username=options.ilo_user, password=options.ilo_pass)
        # Login with the Redfish client
        redfish_obj.login()
    except ServerDownOrUnreachableError as excp:
        sys.stderr.write("ERROR: server not reachable or does not support RedFish.\n")
        sys.exit()
    (ilogen,_) = get_gen(redfish_obj)
    print ("Generation is ", ilogen)
    if int(ilogen) == 5:
        get_SmartArray_LogicalDrives(redfish_obj)
    else:
        get_SmartArray_LogicalDrives_gen9(redfish_obj)
    redfish_obj.logout()
```
